# WRITER

```python
# Combine all metrics DataFrames into a single consolidated DataFrame
# Uses unionByName to merge, preserving column order and names
from functools import reduce
from pyspark.sql import functions as F

all_metrics = [
    metrics_customers,
    metrics_orders,
    metrics_order_items,
    metrics_products,
    metrics_payments,
    metrics_RI
]

consolidated_metrics_df = reduce(lambda df1, df2: df1.unionByName(df2), all_metrics)
# Display the consolidated metrics summary for review
display(consolidated_metrics_df)
```

```python
# Persist the consolidated metrics DataFrame to a Delta table for reporting and monitoring
consolidated_metrics_df.write.mode("overwrite").format("delta").saveAsTable("retail_catalog.default.quality_metrics")
```
