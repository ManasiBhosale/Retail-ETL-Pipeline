```python
# Import standard libraries for directory operations and regular expressions
import os
import re

# Directory containing CSV files to be ingested
csv_dir = "/Volumes/retail_catalog/default/raw_data/"

# Target schema for bronze layer tables
target_schema = "retail_catalog.bronze"

# List all files in the directory and filter for those ending with '.csv'
csv_files = [f for f in os.listdir(csv_dir) if f.lower().endswith('.csv')]

for csv_file in csv_files:
    # Extract table name (remove .csv extension, lowercase for consistency)
    table_name = re.sub(r"\.csv$", "", csv_file, flags=re.IGNORECASE).lower()
    full_table_name = f"{target_schema}.{table_name}"
    file_path = os.path.join(csv_dir, csv_file)
    
    # Read CSV file as Spark DataFrame with header and automatic schema inference
    df = spark.read.csv(file_path, header=True, inferSchema=True)
    
    # Write DataFrame to Delta as a new bronze layer table, overwriting if exists
    df.write.mode("overwrite").saveAsTable(full_table_name)
    print(f"Ingested {csv_file} as table {full_table_name}")
```
