```python
from pyspark.sql.types import StructType, StructField, StringType, TimestampType, IntegerType

customers_schema = StructType([
    StructField("customer_id", StringType(), nullable=False),
    StructField("customer_zip_code_prefix", IntegerType(), nullable=False),
    StructField("customer_city", StringType(), nullable=False),
    StructField("customer_state", StringType(), nullable=False)
])
```

```python
df_customers = (
    spark.read
    .option("header", "true")
    .schema(customers_schema)
    .csv("/Volumes/workspace/retail_schema/raw/Customers.csv")
)
```

```python
df_customers.show(5)
```

```python
df_customers.printSchema()
```

```python
from pyspark.sql.functions import col

df_customers = df_customers.filter(
    col("customer_id").isNotNull()
)
```

```python
df_customers.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/retail_schema/staging/Customers")
```
