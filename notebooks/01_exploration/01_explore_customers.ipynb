# CUSTOMER DATA EXPLORATION

```python
# Import essential libraries for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load customer data from CSV file into a DataFrame
# Show first 5 rows to preview
df_customers = pd.read_csv('/Volumes/retail_catalog/default/raw_data/Customers.csv')
df_customers.head()
```

```python
# Display the shape (number of rows and columns) of the DataFrame
df_customers.shape
```

```python
# Print column info, data types, and memory usage
df_customers.info()
```

```python
# Display summary statistics for numerical columns in the DataFrame
df_customers.describe().T
```

```python
# Count missing values (nulls) in each column
df_customers.isnull().sum()
```

```python
# Count the number of unique values in each column
df_customers.nunique()
```

```python
# Group by state and count unique cities for each state
city_state_counts = df_customers.groupby('customer_state')['customer_city'].nunique()

# Plot a bar chart showing the number of unique cities per state
plt.figure(figsize=(10, 6))
plt.bar(city_state_counts.index, city_state_counts.values)
plt.xticks(rotation=90)
plt.title('Unique Cities per State')
plt.xlabel('State')
plt.ylabel('Number of Unique Cities')
plt.show()
```

```python
# Group by state and count unique customers for each state
customer_state_counts = df_customers.groupby('customer_state')['customer_id'].nunique()

# Plot a bar chart showing the number of unique customers per state
plt.figure(figsize=(10, 6))
plt.bar(customer_state_counts.index, customer_state_counts.values)
plt.xticks(rotation=90)
plt.title('Unique Customers per State')
plt.xlabel('State')
plt.ylabel('Number of Unique Customers')
plt.show()
```
