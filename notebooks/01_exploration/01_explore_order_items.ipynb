```python
plt.figure(figsize=(8, 4))
plt.hist(df_order_items['price'], bins=50)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df_order_items['shipping_charges'], bins=50)
plt.title('Shipping Charges Histogram')
plt.xlabel('Shipping Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.boxplot(df_order_items['price'])
plt.title('Price Distribution')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(8, 5))
plt.boxplot(df_order_items['shipping_charges'])
plt.title('Shipping Charges Distribution')
plt.ylabel('Shipping Charges')
plt.show()
```
