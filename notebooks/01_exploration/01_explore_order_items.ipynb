# ORDER ITEMS DATA EXPLORATION

```python
# Import essential libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load order items data from CSV into a pandas DataFrame
# Preview the first five rows to check successful loading
df_order_items = pd.read_csv('/Volumes/retail_catalog/default/raw_data/OrderItems.csv')
df_order_items.head()
```

```python
# Display the number of rows and columns in the DataFrame
df_order_items.shape
```

```python
# Show information on columns, data types, and missing values
df_order_items.info()
```

```python
# Display summary statistics for all numeric columns
df_order_items.describe().T
```

```python
# Count missing (null) values in each column
df_order_items.isnull().sum()
```

```python
# Count unique values per column to evaluate categorical diversity
df_order_items.nunique()
```

```python
# Visualize the distribution of order item prices using a histogram
plt.figure(figsize=(8, 4))
plt.hist(df_order_items['price'], bins=50)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of freight (shipping charges) with a histogram
plt.figure(figsize=(8, 4))
plt.hist(df_order_items['freight_value'], bins=50)
plt.title('Shipping Charges Histogram')
plt.xlabel('Shipping Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the boxplot of price to spot outliers and distribution shape
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.boxplot(df_order_items['price'])
plt.title('Price Distribution')
plt.ylabel('Price')
plt.show()

# Visualize the boxplot of shipping charges to spot outliers
plt.figure(figsize=(8, 5))
plt.boxplot(df_order_items['freight_value'])
plt.title('Shipping Charges Distribution')
plt.ylabel('Shipping Charges')
plt.show()
```
