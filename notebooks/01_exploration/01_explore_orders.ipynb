```python
cols_to_convert = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_timestamp',
    'order_estimated_delivery_date'
]

df_orders[cols_to_convert] = df_orders[cols_to_convert].apply(
    lambda col: pd.to_datetime(col, dayfirst=True, errors='coerce')
)
```

```python
df_orders.info()
```

```python
min_purchase_date = df_orders['order_purchase_timestamp'].min()
max_purchase_date = df_orders['order_purchase_timestamp'].max()

print('Order Purchase Date Range: {} to {}'.format(min_purchase_date, max_purchase_date))
```

```python
min_approval_date = df_orders['order_approved_at'].min()
max_approval_date = df_orders['order_approved_at'].max()

print('Order Approval Date Range: {} to {}'.format(min_approval_date, max_approval_date))
```

```python
min_delivery_date = df_orders['order_delivered_timestamp'].min()
max_delivery_date = df_orders['order_delivered_timestamp'].max()

print('Order Delivery Date Range: {} to {}'.format(min_delivery_date, max_delivery_date))
```
