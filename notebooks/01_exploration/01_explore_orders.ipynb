# ORDERS DATA EXPLORATION

```python
# Import essential libraries for data handling and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load order data from CSV into a DataFrame
# Preview the first five rows to verify successful loading
df_orders = pd.read_csv('/Volumes/retail_catalog/default/raw_data/Orders.csv')
df_orders.head()
```

```python
# Display the number of rows and columns in the DataFrame
df_orders.shape
```

```python
# Show information on columns, data types, and missing values
df_orders.info()
```

```python
# Display summary statistics for numeric columns
df_orders.describe().T
```

```python
# Count missing (null) values in each column
df_orders.isnull().sum()
```

```python
# Count unique values per column to evaluate categorical diversity
df_orders.nunique()
```

```python
# Convert string columns to datetime format for temporal analysis
cols_to_convert = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

df_orders[cols_to_convert] = df_orders[cols_to_convert].apply(
    lambda col: pd.to_datetime(col, dayfirst=True, errors='coerce')
)
```

```python
# Check column data types after datetime conversion to confirm transformation
df_orders.info()
```

```python
# Find and print the minimum and maximum order purchase dates
min_purchase_date = df_orders['order_purchase_timestamp'].min()
max_purchase_date = df_orders['order_purchase_timestamp'].max()

print('Order Purchase Date Range: {} to {}'.format(min_purchase_date, max_purchase_date))
```

```python
# Find and print the minimum and maximum order approval dates
min_approval_date = df_orders['order_approved_at'].min()
max_approval_date = df_orders['order_approved_at'].max()

print('Order Approval Date Range: {} to {}'.format(min_approval_date, max_approval_date))
```

```python
# Find and print the minimum and maximum order delivery carrier dates
min_delivery_date = df_orders['order_delivered_carrier_date'].min()
max_delivery_date = df_orders['order_delivered_carrier_date'].max()
print('Order Delivery Carrier Date Range: {} to {}'.format(min_delivery_date, max_delivery_date))
```

```python
# Find and print the minimum and maximum order delivery customer dates
min_delivery_date = df_orders['order_delivered_customer_date'].min()
max_delivery_date = df_orders['order_delivered_customer_date'].max()
print('Order Delivery Customer Date Range: {} to {}'.format(min_delivery_date, max_delivery_date))
```
